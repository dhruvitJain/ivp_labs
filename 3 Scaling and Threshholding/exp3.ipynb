```python
from skimage import io
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
image = io.imread ('C:/Users/dhruv/Desktop/College/NOTES/IVP/Labs/lenna_grey.jpg')
io.imshow(image)
```

```python
def scale (x,a):
    y=x+a  #for scaling we multiply, for brightness we add
    return y

image_scale = image.copy()
sh=image_scale.shape
print (sh)
```

```python
for row in range (0,sh[0]):
    for col in range (0,sh[1]):
        image_scale[row][col]= scale (image[row][col],-50)
        
plt.figure()
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
io.imshow (image)
plt.title ('Original Image')

plt.subplot(1,2,2)
io.imshow (image_scale)
plt.title ('Brightness modified image')
```

```python
#thresholding without background
image_gray = rgb2gray (image) #so that we dont have to work in 3 different planes
sh= image_gray.shape #output will be (225,225)
th = int (input("Enter Threshold Value: "))
th
#image_gray rgb2gray will normalise values between 0 and 1 therefore we multiply by 255
image_gray = 255*image_gray
image_gray
```

```python
image_th = image_gray.copy()
for row in range (0,sh[0]):
    for col in range (0,sh[1]):
        if image_gray[row][col] >= th:
            image_th[row][col] = 255
        else:
            image_th [row][col] = 0
            
plt.figure()
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
io.imshow (image_gray, cmap='gray')
plt.title ('Original Image')

plt.subplot(1,2,2)
io.imshow (image_th,cmap='gray')
plt.title ('Threshold image without background')
```

```python
#thresholding with background
image_gray = rgb2gray (image) #so that we dont have to work in 3 different planes
sh= image_gray.shape #output will be (225,225)
th = int (input("Enter Threshold Value: "))
th
#image_gray rgb2gray will normalise values between 0 and 1 therefore we multiply by 255
image_gray = 255*image_gray
image_gray
```

```python
image_th = image_gray.copy()
for row in range (0,sh[0]):
    for col in range (0,sh[1]):
        if image_gray[row][col] >= th:
            image_th[row][col] = 255
        else:
            pass
            
plt.figure()
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
io.imshow (image_gray, cmap='gray')
plt.title ('Original Image')

plt.subplot(1,2,2)
io.imshow (image_th,cmap='gray')
plt.title ('Threshold image with background')
```
