```python
#exp 7 - filter the image 

from skimage import io
import numpy as np
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
```

```python
image_ori=io.imread('home.jpeg')
image_gray=rgb2gray(image_ori)
sh= image_gray.shape
sh
```

```python
#Centering the image then FFT
row=sh[0]
col=sh[1]
for r in range (0,row):
    for c in range(0,col):
        image_gray[r][c]=((-1)**(r+c))*image_gray[r][c]
        
im_fft=np.fft.fft2(image_gray)
```

```python
#construct low pass filter in fft dowmain
low_pass = image_gray.copy()
cent_row = int (sh[0]/2)
cent_col = int (sh[1]/2)

#cutoff freq for lpf = 5 ,20
cutoff_freq = sh[0]/20

#for high pass filter, change if else

for r in range (0,row):
    for c in range (0,col):
        dist = np.sqrt (((r-cent_row)**2)+((c-cent_col)**2))
        if dist > cutoff_freq:
            low_pass [r][c]=0  #for HPF, this is 1
        else:
            low_pass [r][c]=1  #for HPF, this is 0

            
            
#filter the image
image_filtered = im_fft*low_pass

#reconstructing the filtered image i.e spatial domain
image_sp = np.fft.ifft2 (image_filtered)
image_sp = np.abs (image_sp)

plt.figure(figsize=(7,7))
plt.subplot (1,2,1)
io.imshow (image)
plt.title("Original Image")

plt.subplot (1,2,2)
im_log = np.log10(1+np.abs(im_fft))
io.imshow(im_log, cmap='gray')
plt.title("FFT of Image")

plt.figure(figsize=(7,7))
plt.subplot (1,2,1)
io.imshow (low_pass,cmap='gray')
plt.title("Filter")

plt.subplot (1,2,2)
io.imshow (image_sp, cmap='gray')
plt.title("Reconstructed Image")
```
