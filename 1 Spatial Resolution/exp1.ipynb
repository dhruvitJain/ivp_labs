```python
#experiment number 1
#grey/intensity level resolution
from skimage import data
from skimage import io
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
```

```python
image = data.astronaut()
#print (image)
#io.imshow(image)
```

```python
img_gray = rgb2gray (image)
io.imshow(img_gray)
#print (img_gray)
img_gray = img_gray * 255
#print (img_gray)
```

```python
img_reso = img_gray.copy()
b= 3
num_levels = 2**b
step_size = 256//num_levels
print (step_size)
sh = img_reso.shape
print (sh)
```

```python
for r in range (0,sh[0]):
    for c in range (0,sh[1]):
        temp = img_reso[r][c]
        for i in range (0,num_levels):
            if temp >= i*step_size and temp < (i+1)*step_size:
                img_reso [r][c]=i*step_size
                break

        
```

```python
plt.figure()
plt.figure(figsize=(7,7))

plt.subplot (2,1,1)
io.imshow (img_gray, cmap='gray')
plt.title ('Original Image')

plt.subplot (2,1,2)
io.imshow (img_reso, cmap='gray')
plt.title ('Low Reso Image')
```
