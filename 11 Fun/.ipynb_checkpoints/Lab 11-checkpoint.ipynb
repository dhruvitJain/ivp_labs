```python
#blending
from skimage.transform import resize
import numpy as np

im1= io.imread ('dhoni.jpg')
im1=im1/255
im2= io.imread ('ViratKohli.jpg')
im2= im2/255
sh = im2.shape
im1 = resize(im1,(sh[0],sh[1],3))
```
