```python
# experiment 2
# downsample the image
from skimage import io
import matplotlib.pyplot as plt
```

```python
image = io.imread ("C:/Users/dhruv/Desktop/face.png")
io.imshow (image) 
```

```python
img_down = image.copy ()
sh = image.shape
print (sh)
```

```python
ds = 5
rowd = int(sh[0]/ds) +1 
cold = int(sh[1]/ds) +1
img_down = img_down[0:rowd,0:cold]
#sh = img_down.shape
sh = image.shape
print (sh)
```

```python
rw=0 
cl=0

for r in range (0,sh[0],ds):
    cl=0
    for c in range (0,sh[1],ds):
        img_down [rw][cl]= image[r][c]
        cl += 1
    rw += 1
    
sh1 = img_down.shape
print (sh1)
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
io.imshow (image)
plt.title ('Original Image')

plt.subplot(1,2,2)
io.imshow (img_down)
plt.title ('Down sampled Image')
```

```python
up = 5
img_up = image.copy()
sh = img_down.shape

r=0
c=0

for row in range (0,sh[0]):
    c=0
    for col in range (0,sh[1]):
        img_up[r][c]=img_down[row][col]
        img_up [r,c+1:c+up] = 0
        c+=up
    img_up [r+1 : r+up, :] = 0
    r+=up
        
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
io.imshow (image)
plt.title ('Original Image')

plt.subplot(1,3,2)
io.imshow (img_down)
plt.title ('Down sampled Image')

plt.subplot(1,3,3)
io.imshow (img_up)
plt.title ('Up sampled Image')
```

```python
shu = img_up.shape
#use average method of interpolation
#for columns
for row in range (0,shu[0],up):
    for col in range (0,shu[1]-up,up):
        temp1 = img_up[row,col]/2
        temp2 = img_up[row,col+up]/2
        temp = temp1 + temp2
        img_up [row,col+1 : col+up] = [int(temp[0]),int(temp[1]),int(temp[2])] 
        
#for rows        
for row in range (0,shu[0]-up,up):
    for col in range (0,shu[1]):
        temp1 = img_up[row,col]/2
        temp2 = img_up[row+up,col]/2
        temp = temp1 + temp2
        img_up [row:row+up,col] = [int(temp[0]),int(temp[1]),int(temp[2])]      
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
io.imshow (image)
plt.title ('Original Image')

plt.subplot(1,3,2)
io.imshow (img_down)
plt.title ('Down sampled Image')

plt.subplot(1,3,3)
io.imshow (img_up)
plt.title ('Up sampled Image')
```
